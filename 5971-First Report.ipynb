## Abstract

Large Language Models (LLMs) have demonstrated strong potential in question answering and decision support. However, their abilities in specific domains remains limited due to hallucination, lack of transparency, and insufficient grounding evidence. For a high-stakes field like healthcare, a generic "one-size-fits-all" model fails to provide the precision and reliability required for clinical decision support. For this project, we propose a QA system that designed to bridge the gap between raw LLM capabilities and the requirements of professional medical consultation, to provide accurate and evidence-based clinical support. 

## Introduction & Description

Online medical information is abundant, yet it is often inconsistent or difficult for non-expert users to interpret. When people search for symptom explanations, they usually faces some challanges: overly simplified advice that lacks context, or highly technical content that is not accessible. Even though LLMs have made it easier to ask questions in natural language and obtain information quickly, in information-dense domain such as healthcare settings, simply generated statements from pre-trained LLMs alone is not enough, and hallucinated medical claims can mislead users can be harmful (Ji et al., 2023; Singhal et al., 2023). This creates a clear gap between the convenience of LLM-based interaction and the requirements of high-stakes decision-making.

Motivated by these challanges, our project focuses on developing an analytical framework for reliable, patient-centered medical QA system that emphasizes evidence support and decision relevance. The societal value of improving these decisions is substantial: more reliable medical guidance can reduce misinformation and encourage earlier interventions when necessary (Singhal et al., 2023). In addition, improved decision support has practical value by reducing unnecessary clinic visits, easing the informational burden on providers, and promoting more effective communication between patients and clinicians (Lizée et al., 2024). We believe such a system can help shape everyday healthcare decisions, including whether an individual should seek professional medical attention or manage symptoms at home, and what appropriate next steps or follow-up questions may be needed. These types of decisions arise at scale on a daily basis and have direct implications for the effective use of medical resources.

This project is important to our team for two main reasons. First, we believe that the potential of LLMs extends far beyond their current general-purpose applications. By adapting and fine-tuning these models for knowledge-intensive domains, we aim to explore how their capabilities can be fully leveraged in specialized settings such as healthcare. Second, access to medical knowledge remains highly uneven across different communities, largely due to disparities in healthcare resources and education. Many individuals face significant barriers when trying to obtain trustworthy medical information or understand professional terminology. Through this project, we aim to narrow this gap by providing a low-barrier, patient-friendly system that enables broader access to evidence-based medical guidance. Ultimately, our goal is to support more equitable healthcare understanding and informed decision-making through responsible artifical intelligence.

## Methodology

Due to the project involves multiple alignment stages, different types of datasets will be reilied on throughout the fine-tuning pipeline. In particular, supervised fine tuning (SFT) and preference-based alignment methods such as RLHF or DPO require distinct data assets, as they optimize the model for different objectives. Instruction-tuning datasets focus on teaching the model to generate clinically appropriate responses to user queries, while preference datasets are designed to further improve response quality through chosen–rejected comparisons.

Accordingly, our data strategy is organized by training stage. For the supervised fine-tuning phase, we plan to leverage medical dialogue and question-answer resources, which provides patient-style prompts paired with responses. For the alignment stage, we will explore preference-based datasets such as *full-hh-rlhf* or *DPO-En-Zh-20k-Preference* which contain ranked response pairs that can be used to discourage unsafe or low-quality outputs. **Data Assets Used in Pipeline**   

For the SFT stage, we plan to leverage the *ChatDoctor* released by Li et al. The dataset consists of real-world doctor–patient conversational data collected from online medical consultation platforms. According to the project documentation, the training corpus includes approximately 100,000 medical dialogues from HealthCareMagic, an additional 10,000 conversations from iCliniq, and several generated physician–patient interactions designed to enrich disease coverage. These dialogues span a range of common symptoms, diagnoses, medication inquiries, and follow-up recommendations. The conversational and patient-facing structure of ChatDoctor makes it highly suitable for our intended clinical question answering scenario. By fine-tuning on this dataset, the model can learn domain-relevant response patterns and improve its ability to provide understandable explanations and actionable guidance. From an analytics perspective, ChatDoctor directly supports decision-relevant outputs, such as helping users determine whether professional care is needed, interpret unfamiliar medical terminology, and identify appropriate next steps in response to health concerns.

For the RLHF stage, two datasets are intended to be used: *Dahoas/full-hh-rlhf* and *OpenAssistant Conversations (OASST1)*. The Dahoas/full-hh-rlhf dataset contains 110,000 human preference comparisons derived from Anthropic's Helpful and Harmless data, which includes ranked response pairs designed to optimize helpfulness while minimizing harmful outputs. This dataset provides structured feedback on conversational quality, safety, and informativeness, making it valuable for aligning the model with user expectations in sensitive domains such as healthcare. The OASST1 dataset, collected through a crowdsourced effort, offers multilingual conversational trees with human ratings that reflect preferences for clarity, accuracy, and engagement. These annotations span diverse types and styles, enabling the model to learn distinctions between high-quality and suboptimal responses. By training on these datasets through reinforcement learning from human feedback, the model can better balance medical accuracy with conversational naturalness and user safety. From an alignment perspective, these datasets help ensure that the fine-tuned model not only provides clinically relevant information, but also communicates in a empathetic manner.


**Analytical Framework and Model Evaluation**

Our project adopts analytical frameworks that includes descriptive, predictive, and prescriptive to support reliable decision-making. This framework is helpful to evaluate if the responces are accurate or actionable.

In terms of descriptive analytics, we plan to characterize patterns within the training data and model outputs. This includes analyzing the distribution of symptom types, and patient concerns across the dataset to ensure adequate coverage of health issues. We also conduct analysis of response quality, examining factors such as length and terminology complexity, and the presence of evidence citations. These insights helps to understand baseline model behavior and identify areas where additional training may be needed. 

The predictive analytics focuses on forecasting decision-relevant outcomes based on user queries, and is mainly implemented through the LLM's ability to perform probabilistic inference. By analyzing the information from users, the model predicts the most likely clinical diagnoses and assesses potential health risks. We evaluate the performance of these predictions using metrics such as Top-k Accuracy, ensuring that the model's predictions match the consensus of medical experts.

The prescriptive analytics mainly provides recommendations to guide patient decision-making. Based on the LLMs' understanding of clinical patterns, the system suggests appropriate next steps, such as scheduling an appointment or seeking emergency care. Prescriptive outputs also include patient educations such as medicine explanation. 

**Method Novelty**

The key aspect of the project is its patient-centered framing. Much of the existing literature on medical large language models has primarily focused on specialized clinical reasoning. In contrast, fewer systems are designed for patients who usually lack of medical information. Therefore, emphasizes data selection and system design from the patient’s perspective is more important. We will prioritize datasets that reflect real patient inquiries and  explanations, and we aim to generate responses that are not only medically relevant but also clear and user-friendly. By focusing on communication quality and decision relevance for everyday users, our work explores how medical LLMs can better support health literacy and reduce barriers to medical knowledge.



## Related Work

**Medical LLMs for Clinical Question Answering**

Recent work shows that LLMs can encode substantial clinical knowledge and perform strongly on medical QA benchmarks. Singhal et al., from Google, intoduced *Med-PaLM line of research*, which evaluates LLMs on MultiMedQA and emphasizes clinician-graded axes such as reasoning, potential harm, reflecting a strongly clinician-facing evaluation perspective. (Singhal et al., 2023; Singhal et al., 2025).

Beyond general LLMs adapted to medicine, the community also built domain-specific biomedical language models. Luo et al. introduced BioGPT as a generative transformer that is pre-trained on large biomedical literature, demonstrating strong performance across biomedical NLP tasks and competitive results on biomedical QA benchmarks (Luo et al., 2022). These efforts establish that domain adaptation can improve medical language understanding and generation. These works often prioritize benchmark accuracy while lack of user expeirences.

**Retrieval-Augmented Generation and Evidence Grounding**

One of the challanges to solve is that LLMs producing confident but unsupported statements. Retrieval-Augmented Generation (RAG) addresses this limitation by grounding model outputs in external retrieved documents, enabling knowledge provenance and improved factual consistency (Lewis et al., 2020). 

While RAG was originally introduced as a general framework for knowledge-intensive NLP tasks, it has become highly relevant in biomedical and clinical applications. Recent work has explored RAG for medical question answering, showing its potential to reduce hallucinations and support evidence-based responses (Yuan et al., 2024). In patient-facing healthcare settings, grounding answers in verifiable medical sources is increasingly viewed as a practical strategy to improve trustworthiness and reduce clinical risk (Singhal et al., 2023).


**Preference Alignment in Medical-Domain LLMs**

Several domain-specific efforts have emerged to address this need. Fleming et al. introduced *MedAlign*, a clinician-generated benchmark that incorporates expert preference judgments aligned with real clinical requirements, highlighting that standard metrics often fail to capture medically meaningful response quality (Fleming et al., 2024). The study also demonstrates that clinician rankings provide a more realistic evaluation signal for medical language models than exam-style benchmarks alone. In parallel, alignment techniques such as DPO have been explored as lightweight alternatives to traditional PPO-based RLHF pipelines (Rafailov et al., 2023). Recent clinical research has shown growing interest in applying preference optimization to suppress hallucinations and improve factuality in medical generation tasks (Liu et al., 2025).
